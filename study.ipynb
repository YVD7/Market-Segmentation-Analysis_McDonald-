# Market Segmentation Analysis

```python
# importing essential libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data from source and saving into download.csv file

import requests # importing requests library to request url source page 
req = requests.get("https://homepage.boku.ac.at/leisch/MSA/datasets/mcdonalds.csv") # reqesting url to read data
url_content = req.content  # reading the content from source page
csv_file = open('download.csv','wb') # storing the data into .csv file 
csv_file.write(url_content) # writing the data in download file 
csv_file.close() # now closing the file that been written download.csv
```

```python
# reading downloaded dataset donwload.csv
df = pd.read_csv('download.csv')
df
```

```python
# checking info like data-types, null values, and number of columns and rows, features

df.info()
```

```python
# Checking for count of below variables

df['Gender'].value_counts()
df['VisitFrequency'].value_counts()
df['Like'].value_counts()
```

# Exploring Data

```python
# Exploring data

# Customer segemetation -based on socio-demographic (Age & Genders)

# Gender

labels = ['Female','Male']
size = df['Gender'].value_counts()
colors =['pink','cyan']
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (7,7)
plt.pie(size, colors = colors, explode = explode, labels = labels, shadow = True, autopct='%.2f%%')
plt.title('Gender',fontsize = 20)
plt.axis('off')
plt.legend()
plt.show()

# Age
plt.rcParams['figure.figsize'] = (25, 8)
f = sns.countplot(x = df['Age'], palette='hsv')
f.bar_label(f.containers[0])
plt.title('Age distribution of customers')
plt.show()

# McDonalds recieve more customers of age between 50-60 and 35-40.
```

```python
# Customers segemntation - based on pyschographic segementation

# For convinence remainig the category
df['Like'] = df['Like'].replace({'I hate it!-5':'-5','I love it!+5':'+5'})

# Like
sns.catplot(x="Like",y = "Age", data= df, 
            orient="v", height=5, aspect=2, palette="Set2", kind="swarm")

plt.show('Likelyness of McDonald w.r.t Age')
plt.show
```

```python
# Label encoding of categorical - Converting 11 cols with yes/no

from sklearn.preprocessing import LabelEncoder
def labelling(x):
    df[x] = LabelEncoder().fit_transform(df[x])
    return df

cat = ['yummy', 'convenient','spicy','fattening','greasy','fast','cheap',
        'tasty','expensive','healthy','disgusting']

for i in cat:
    labelling(i)
df
```

```python
# Histogram of the each attibutes

plt.rcParams['figure.figsize'] = (12, 15)
df.hist()
plt.show()
```

```python
# considering only first 11 attributes
df_eleven = df.loc[:,cat]
df_eleven
```

```python
# Considering only the 11 cols and converting it into array

x = df.loc[:, cat].values
x
```

```python
# Principal component analysis

from sklearn.decomposition import PCA
from sklearn import preprocessing

pca_data = preprocessing.scale(x)

pca = PCA(n_components=11)
pc= pca.fit_transform(x)

names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11']
pf = pd.DataFrame(data=pc, columns= names)
pf
```

```python
# Proportion of Variance (from PC1 to PC 11)

pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
# correlation coefficient between original variables and the components

loadings = pca.components_
num_pc = pca.n_features_
pc_list = ['pc'+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = df_eleven.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
# correlation matrix plot for loadings

plt.rcParams['figure.figsize'] = (20, 25)
ax = sns.heatmap(loadings_df, annot=True, cmap='Spectral')
plt.show()
```

**EXTRACTING SEGMENTS**

```python
# Extracting segments

# Using k-means clustering analysis

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(df_eleven)
visualizer.show()
```

```python
# kmeans clustering 

kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df_eleven)
df['cluster_num'] = kmeans.labels_ # adding to df
print(kmeans.labels_) #Label assigned for each data point
print(kmeans.inertia_) # gives within-cluster sum of squares
print(kmeans.n_iter_) #number of iterations that k-means algorthms run to get a minimum within-cluster sum squares
print(kmeans.cluster_centers_) # Location of centroid on each cluster.
```

```python
# To see each cluster size
from collections import Counter
Counter(kmeans.labels_)
```

```python
# Visualizing clusters
sns.scatterplot(data=pf, x="pc1", y="pc2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            marker="X",c="r",s=80, label="centroids")
plt.legend()
plt.show()
```

**Selecting target semgent**

```python
# Calculating the mean 
# visit frequency

df['VisitFrequency'] = LabelEncoder().fit_transform(df['VisitFrequency'])
visit  = df.groupby('cluster_num')['VisitFrequency'].mean()
visit = visit.to_frame().reset_index()
visit
```

```python
# Like 
df['Like'] = LabelEncoder().fit_transform(df['Like'])
Like  = df.groupby('cluster_num')['Like'].mean()
Like = Like.to_frame().reset_index()
Like
```

```python
# Gender
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
Gender = df.groupby('cluster_num')['Gender'].mean()
Gender = Gender.to_frame().reset_index()
Gender
```

```python
segment = Gender.merge(Like, on='cluster_num', how='left').merge(visit, on = 'cluster_num',how='left')
segment
```

```python
# Target segments

plt.figure(figsize=(9,4))
sns.scatterplot(x="VisitFrequency",y="Like",data=segment, s=400, color='r')
plt.title("Sample segment evaluation plot for the fast food data set", fontsize = 15)
plt.xlabel("Visit", fontsize = 12)
plt.ylabel("Like", fontsize = 12)
plt.show()
```
